```python
import yfinance as yf
import pandas as pd
from datetime import datetime, timedelta
import matplotlib.pyplot as plt
import plotly.graph_objects as go
from datetime import datetime, timedelta
```

```python
# Definir os tickers dos ativos
tickers = ['AAPL', 'GOOGL', 'MSFT']
```

```python
# Definir o período
end_date = datetime.now()
start_date = end_date - timedelta(days=2*365)
```

```python
# Baixar os dados
for ticker in tickers:
    data = yf.download(ticker, start=start_date, end=end_date)

    # Salvar os dados em um arquivo CSV
    filename = f"{ticker}_data.csv"
    data.to_csv(filename)
    print(f"Dados de {ticker} salvos em {filename}")

print("Download e salvamento concluídos.")
```

```python
# Lista de tickers
tickers = ['AAPL', 'GOOGL', 'MSFT']

# Dicionário para armazenar os DataFrames
dfs = {}
```

```python
# Carregar dados e calcular amplitude
for ticker in tickers:
    # Carregar dados do CSV
    df = pd.read_csv(f"{ticker}_data.csv", index_col='Date', parse_dates=True)

    # Calcular amplitude
    df['Amplitude'] = df['High'] - df['Low']

    # Armazenar DataFrame no dicionário
    dfs[ticker] = df
```

```python
# Criar gráfico
plt.figure(figsize=(12, 6))

for ticker, df in dfs.items():
    plt.plot(df.index, df['Amplitude'], label=ticker)

plt.title('Amplitude de Variação Diária por Ativo')
plt.xlabel('Data')
plt.ylabel('Amplitude (Máximo - Mínimo)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Calcular volatilidade (desvio padrão da amplitude)
volatilidades = {ticker: df['Amplitude'].std() for ticker, df in dfs.items()}

# Identificar ativo com maior volatilidade
ativo_mais_volatil = max(volatilidades, key=volatilidades.get)

print("Volatilidade por ativo:")
for ticker, volatilidade in volatilidades.items():
    print(f"{ticker}: {volatilidade:.2f}")

print(f"\nAtivo com maior volatilidade: {ativo_mais_volatil}")
```

```python
# Carregar dados, calcular média móvel e desvio padrão
for ticker in tickers:
    # Carregar dados do CSV
    df = pd.read_csv(f"{ticker}_data.csv", index_col='Date', parse_dates=True)

    # Calcular média móvel de 90 dias do fechamento ajustado
    df['MA_90'] = df['Adj Close'].rolling(window=90).mean()

    # Calcular desvio padrão móvel de 90 dias do fechamento ajustado
    df['STD_90'] = df['Adj Close'].rolling(window=90).std()

    # Armazenar DataFrame no dicionário
    dfs[ticker] = df
```

```python
# Criar gráfico de médias móveis
plt.figure(figsize=(12, 6))
for ticker, df in dfs.items():
    plt.plot(df.index, df['MA_90'], label=ticker)

plt.title('Média Móvel de 90 dias do Fechamento Ajustado')
plt.xlabel('Data')
plt.ylabel('Preço')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Criar gráfico de desvios padrão móveis
plt.figure(figsize=(12, 6))
for ticker, df in dfs.items():
    plt.plot(df.index, df['STD_90'], label=ticker)

plt.title('Desvio Padrão Móvel de 90 dias do Fechamento Ajustado')
plt.xlabel('Data')
plt.ylabel('Desvio Padrão')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Calcular volatilidade total (desvio padrão do fechamento ajustado)
volatilidades = {ticker: df['Adj Close'].std() for ticker, df in dfs.items()}

# Identificar ativo com maior volatilidade
ativo_mais_volatil = max(volatilidades, key=volatilidades.get)

print("Volatilidade total por ativo:")
for ticker, volatilidade in volatilidades.items():
    print(f"{ticker}: {volatilidade:.2f}")

print(f"\nAtivo com maior volatilidade total: {ativo_mais_volatil}")
```

```python
# Carregar dados do CSV
df = pd.read_csv(f"{ticker}_data.csv", index_col='Date', parse_dates=True)
```

```python
# Selecionar os últimos 60 dias
end_date = df.index[-1]
start_date = end_date - timedelta(days=60)
df_60d = df.loc[start_date:end_date]
```

```python
# Criar o gráfico de candlestick
fig = go.Figure(data=[go.Candlestick(x=df_60d.index,
                open=df_60d['Open'],
                high=df_60d['High'],
                low=df_60d['Low'],
                close=df_60d['Close'])])
```

```python
# Personalizar o layout
fig.update_layout(
    title=f'{ticker} - Últimos 60 dias',
    yaxis_title='Preço',
    xaxis_title='Data',
    xaxis_rangeslider_visible=False
)

# Mostrar o gráfico
fig.show()
```
